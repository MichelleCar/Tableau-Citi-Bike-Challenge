```python
# Dependencies
import pandas as pd
import datetime
```

```python
# List of file names to be combined 
files = ['data/202201.csv', 
         'data/202202.csv', 
         'data/202301.csv',
         'data/202302.csv']

# Create an empty list to store each dataframe
df_list = []

# Loop through each file in the list
for file in files:
    # Read the file into a dataframe
    df = pd.read_csv(file)

    # Append the dataframe to the list
    df_list.append(df)

# Concatenate all dataframes in the list into one dataframe
df = pd.concat(df_list)

# Show dataframe
df
```

```python
# Verify the type of data 
df.info()
```

```python
# Convert "started_at", "ended_at" from object to date/time
df[["started_at", "ended_at"]] = df[["started_at", "ended_at"]].apply(pd.to_datetime)
df
```

```python
# Calculate duration of each trip
df['trip_duration'] = df['ended_at'] - df['started_at']
df
```

```python
# Remove all rows where duration of trip is less than 30 minutes
df_filtered = df[df['trip_duration'] >= "0 days 00:30:00"]
df_filtered
```

```python
# Remove all rows with "null" data
df_filtered = df_filtered.dropna(how='any',axis=0)
df_filtered
```

```python
# Save the dataframe to a CSV file
df_filtered.to_csv('data/citibike.csv', index=False)
```
